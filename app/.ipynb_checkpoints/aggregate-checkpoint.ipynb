```python
import boto3 

s3 = boto3.client('s3',
    endpoint_url = 'http://minio:9000',
    aws_access_key_id = 'admin',
    aws_secret_access_key = 'admin123'
)

s3
```

```python
import pandas as pd
import io
from datetime import datetime

res = s3.get_object(Bucket='bronze',Key='netflix_titles.csv')
text = io.BytesIO(res['Body'].read())

df = pd.read_csv(text)
```

```python
df = df.dropna().reset_index()
```

```python
df['processed_at'] = datetime.now()
```

```python
df.columns
```

```python
df['explode_list'] = df['listed_in'].str.split(',').apply(lambda x:list(map(lambda y:y.strip(),x)))
```

```python
df = df.explode('explode_list')
```

```python
df
```

```python
output=io.BytesIO()
df.to_parquet(output,engine='pyarrow',index=False)
output.seek(0)
```

```python
s3.put_object(
    Bucket='silver',
    Key='silver.parquet',
    Body=output.getvalue()
)
```

```python
res = s3.get_object(Bucket='silver',Key='silver.parquet')
```

```python
res['ResponseMetadata']['HTTPHeaders']['content-length']#['contentLength']
```

```python
data = res['Body'].read()
```

```python
gold = pd.read_parquet(io.BytesIO(data), engine="pyarrow")
gold
```

# Aggregate

```python
df = pd.DataFrame(gold['explode_list'].value_counts()).reset_index()
```

```python
output=io.BytesIO()
df.to_parquet(output,engine='pyarrow',index=False)
output.seek(0)

s3.put_object(
    Bucket='gold',
    Key='gold.parquet',
    Body=output.getvalue()
)
```

```python
res = s3.get_object(Bucket='gold',Key='gold.parquet')
```
